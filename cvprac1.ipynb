```python
# Define the scale factor
# Increase the size by 3 times
scale_factor_1 = 3.0
# Decrease the size by 3 times
scale_factor_2 = 1/5.0
```

```python
new_height=int(height*scale_factor_1)
new_width=int(width*scale_factor_1)
```

```python
new_height1 = int(height * scale_factor_2)
new_width1 = int(width * scale_factor_2)
# Scaled image
scaled_image = cv.resize(src=gray_image ,
                          dsize =(new_width1, new_height1),
                          interpolation=cv.INTER_AREA)
```

```python
# prompt: give me code using im show in image processing for city block pattern using matrices

import numpy as np
import matplotlib.pyplot as plt

# Create a 256x256 image with a black background
image = np.zeros((256, 256), dtype=np.uint8)

# Generate the city block pattern
city_block_pattern = np.zeros((256, 256), dtype=np.uint8)

for i in range(256):
    for j in range(256):
        city_block_pattern[i, j] = abs(i - 128) + abs(j - 128)

# Copy the city block pattern onto the image
image[:] = city_block_pattern

# Display the image using imshow
plt.imshow(image, cmap='gray')
plt.show()

```

```python
# prompt: give me code using im show in image processing for eucledian pattern using matrices

import numpy as np
import cv2

def euclidean_distance(p1, p2):
  return np.sqrt(np.sum(np.square(p1 - p2)))

def generate_euclidean_distance_pattern(rows, cols):
  euclidean_pattern = np.zeros((rows, cols), dtype=np.float32)

  for i in range(rows):
    for j in range(cols):
      euclidean_pattern[i, j] = euclidean_distance(np.array([i, j]), np.array([rows // 2, cols // 2]))

  return euclidean_pattern

# Create a 256x256 image with a black background
image = np.zeros((256, 256), dtype=np.uint8)

# Generate the Euclidean distance pattern
euclidean_pattern = generate_euclidean_distance_pattern(256, 256)

# Copy the Euclidean distance pattern onto the image
image[:] = euclidean_pattern

# Display the image using imshow
cv2_imshow( image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
